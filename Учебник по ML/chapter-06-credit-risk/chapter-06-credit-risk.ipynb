```python
import pandas as pd

import numpy as np

import seaborn as sns

from matplotlib import pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('CreditScoring.csv')
```

```python
df.head()
```

# Очистка данных

```python
df.columns = df.columns.str.lower()
df.head()
```

```python
status_values = {
    1: 'ok',
    2: 'default',
    0: 'unk'
}
```

```python
df.status = df.status.map(status_values)
df.head()
```

```python
home_values = {
    1: 'rent',
    2: 'owner',
    3: 'private',
    4: 'ignore',
    5: 'parents',
    6: 'other',
    0: 'unk'
}

df.home = df.home.map(home_values)
```

```python
marital_values = {
    1: 'single',
    2: 'married',
    3: 'widow',
    4: 'separated',
    5: 'divorced',
    0: 'unk'
}

df.marital = df.marital.map(marital_values)

records_values = {
    1: 'no',
    2: 'yes',
    0: 'unk'
}

df.records = df.records.map(records_values)

job_values = {
    1: 'fixed',
    2: 'parttime',
    3: 'freelance',
    4: 'others',
    0: 'unk'
}

df.job = df.job.map(job_values)
```

```python
df.head()
```

```python
df.describe().round()
```

```python
for c in ['income', 'assets', 'debt']:
    df[c] = df[c].replace(to_replace=99999999, value=np.nan)
```

```python
df.describe().round()
```

```python
df.status.value_counts()
```

```python
df = df[df.status != 'unk']
```
